# Ethics


## Case Study: Data Privacy in Health Care
### Data Ethics Challenge
One significant data ethics challenge is data privacy, particularly in the health care sector. This issue arises when sensitive health information is collected, stored, and shared without adequate safeguards, leading to potential breaches of privacy and trust.

Real World Example
A notable instance of this challenge occurred in 2019 when a major health insurance company experienced a data breach that exposed the personal health information of over 3 million patients. The breach involved unauthorized access to a database that contained sensitive data, including names, Social Security numbers, medical histories, and insurance details.

### Ethical Questions:

Were the patients adequately informed about how their data would be used and protected?
Did the company implement sufficient security measures to safeguard sensitive information?
What are the implications for patient trust in health care providers and insurance companies?
Potential Harms: The breach led to several undesirable consequences, including:

### Identity Theft: Exposed personal information can be exploited for fraudulent activities.
Loss of Trust: Patients may feel hesitant to share their health information, impacting their willingness to seek care or participate in health studies.
Psychological Impact: Patients may experience anxiety and stress over the potential misuse of their sensitive data.
Potential Solutions: To mitigate such risks, the following measures could be implemented:

### Enhanced Security Protocols: Regular audits and updates to security systems to protect sensitive information.
Transparent Data Use Policies: Clear communication with patients about how their data is collected, used, and stored, along with explicit consent forms.
Data Minimization Practices: Collect only the necessary information required for specific purposes and avoid excessive data retention.

# Defining data

### 1. Spreadsheet of Customer Phone Numbers
Structure Type: Structured\
Value Type: Qualitative (phone numbers are not typically used mathematically)\
Source Type: Secondary (the data is received from the parent company)
### 2. Heart Rate Data in JSON Format
Structure Type: Semi-Structured (JSON has a flexible structure but is organized with key-value pairs)\
Value Type: Quantitative (heart rate is a numerical measurement)\
Source Type: Primary (the data is collected directly from the smartwatch wearer)
### 3. Workplace Survey of Employee Morale in a CSV File
Structure Type: Semi-Structured (CSV files)\
Value Type: Qualitative (survey responses may include subjective measures of morale)\
Source Type: Primary (the data is collected directly from employees)
### 4. Database of Galaxies Collected by a Space Probe
Structure Type: Structured (the database is organized, likely in rows and columns)\
Value Type: Quantitative (the number of planets is a numerical measurement)\
Source Type: Secondary (the data has been collected by a space probe and accessed by astrophysicists)
### 5. Personal Finance App Using APIs
Structure Type: Structured (the data is presented in rows and columns)\
Value Type: Quantitative (financial transactions and calculations involve numerical values)\
Source Type: Primary (the app collects data from the user's financial accounts)

```python
df = pd.read_csv("diabetes.csv",sep='\t')
df.head()
```

In this dataset, columns as the following:

Age and sex are self-explanatory\
BMI is body mass index\
BP is average blood pressure\
S1 through S6 are different blood measurements\
Y is the qualitative measure of disease progression over one year\
Let's study this dataset using methods of probability and statistics.

```python
# Compute mean and variance for all columns
mean_values = df.mean()
variance_values = df.var()

print("Mean Values:\n", mean_values)
print("\nVariance Values:\n", variance_values)
```

```python
def plot_boxplot(data, title):
  plt.figure(figsize= (10, 2))
  plt.boxplot(data, vert= False, showmeans= True)
  plt.title(title)
  plt.grid(color= 'gray', linestyle= 'dotted')
  plt.tight_layout()
  plt.show()
```

```python
df_male = df[df['SEX'] == 2]
df_femal = df[df['SEX']== 1]

plot_boxplot(df_male['BMI'], 'BMI of male')
plot_boxplot(df_femal['BMI'], 'BMI of female')

plot_boxplot(df_femal['BP'], 'BP of male')
plot_boxplot(df_femal['BP'], 'BP of female')

plot_boxplot(df_femal['Y'], 'Y of male')
plot_boxplot(df_femal['Y'], 'Y of female')
```

```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)# sai số chuẩn trung bình se = s/ sqrt(n) trong đó s là độ lệch chuẩn mẫu
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m,se, h


keys= ['AGE', 'BMI', 'BP', 'Y']

for key in keys:
  print(f'\nConfidence interval of {key}:')
  for p in [0.85, 0.9, 0.95]:
    m,se, h = mean_confidence_interval(df[key].ffill(),p)
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

```python
# Plot histograms and KDE
plt.figure(figsize=(15, 10))

for i, column in enumerate(['AGE', 'SEX', 'BMI', 'Y']):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Separate the data by gender
male_y = df[df['SEX'] == 2]['Y']
female_y = df[df['SEX'] == 1]['Y']

# Perform t-test
t_stat, p_value = stats.ttest_ind(male_y, female_y)

print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpret the result
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in diabetes progression between men and women.")
else:
    print("Fail to reject the null hypothesis: No significant difference in diabetes progression between men and women.")
```
